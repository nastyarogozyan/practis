```python
import time
import pandas as pd
import requests 
from bs4 import BeautifulSoup as BS
 
 
from urllib3.exceptions import InsecureRequestWarning
from urllib3 import disable_warnings
 
 
df = pd.DataFrame({'Title': [], 'Category': [], 'Tags': [], 'Text': [], 'Link': []})
 
pel = "https://goldapple.ru/makijazh"
r = requests.get(pel)
html = BS(r.content, 'html.parser')
 
list_cat =  ['lico', 'glaza', 'guby', 'brovi', 'nogti', 'kisti']
list_title= []
list_url = []
list_pep = []
list_text = []
list_tags = []
list_tag = []
 
for cat in list_cat:
    page = 0    
    while page!=50:
        disable_warnings(InsecureRequestWarning)
        pel = f"https://goldapple.ru/makijazh/{cat}?p={page}"
        r = requests.get(pel, verify=False)
        html = BS(r.content, 'html.parser')
        items = html.find_all('div', class_='product details product-item-details')
 
        if(len(items)):
            for el in items:
                time.sleep(12)
                title = el.find('div', class_="product-item-category-title")
                title = title.find('span')
 
                item_url = el.find('a', class_="product-item-link js-clamping")
                item_url = item_url.get('href')
 
 
                title = title.text
                #title = title.replace('\n', ' ', ':')
                list_title.append(title)
                list_url.append(item_url)
 
 
                opa = requests.get(item_url, verify=False)
                soap = BS(opa.content, 'html.parser')
 
                pep = soap.find('section', class_='product-description__description')
                pept = pep.text
 
                list_pep.append(pept)
                list_pep.append(" ")
 
                text = " ".join(list_pep)
                #text = text.replace('\xa0', ' ')
                list_text.append(text)
 
                for tag in soap.find_all('span', class_='product-attributes__item-value-text'):
                    list_tag.append(tag.text)
 
                tags = " ".join(list_tag)
                #tags = tags.replace('\n', ' ')
                list_tags.append(tags)
 
                list_pep = []
                list_tag = []
 
 
                df.loc[ len(df.index )] = [title, cat, tags, text, item_url]
 
                print(f"{title}:")
                print(f"{item_url}")
                print("\n")
                print(text)
                print("\n")
                print(tags)
                print("\n")
                print(cat)
                print("page = ", page)
                print("+++++++++++++++++++++++++++++++++++++++++++++")
 
        else: 
            break
        page += 1
  
```

```python
import json
js = json.loads(df.to_json(force_ascii=False, orient='records'))
with open('corpus_brovi21.json', 'w', encoding ='utf-8') as file:
    for i in js:
        file.write(json.dumps(i,ensure_ascii=False) + '\n')
```
